```python
#embedding techniques:convertting our text to vectors


import os
from dotenv import load_dotenv
load_dotenv() #load all the environment varaibles
```

```python
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings # for using openai embedding
embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text="this is tutorial on open ai embedding"
query_result=embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
embeddings_1024=OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
```

```python
text="this is tutorial on open ai embedding"
query_result=embeddings_1024.embed_query(text)
len(query_result)
```

```python
from langchain_community.document_loaders import TextLoader
loader=TextLoader("speech.txt")
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_document=text_splitter.split_documents(docs)
final_document
```

```python
#vector embedding and vector storeDB
from langchain_community.vectorstores import Chroma
db=Chroma.from_documents(final_document,embeddings_1024)
```

```python
db
```

```python
## retrieve the resukt fro query vectorstore db
query="It will be all the easier for us to conduct ourselves as belligerents in a high spirit of right "
retrieved_results=db.similarity_search(query)
print(retrieved_results)
```
